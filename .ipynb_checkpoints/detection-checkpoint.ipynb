```python
"""
Chip Rack Detector — 100% Local, No API Key
--------------------------------------------
Crops regions of a rack photo and matches each region against your
reference chip images using CLIP embeddings (runs on your machine).
 
Usage:
    python detect_chips.py <rack_image>
    python detect_chips.py rack_photo.jpg
    python detect_chips.py rack_photo.jpg --refs ./chip_refs --threshold 0.75
 
Requirements:
    pip install torch torchvision transformers Pillow opencv-python-headless
 
Folder layout:
    chip_refs/
        doritos-cool-ranch.png
        lays-baked.png
        ...
    detect_chips.py
    rack_photo.jpg
"""
 
import sys
import os
import argparse
from pathlib import Path
 
import cv2
import numpy as np
from PIL import Image
import torch
from transformers import CLIPProcessor, CLIPModel
```

```python
# ── Config ─────────────────────────────────────────────────────────────────────
 
DEFAULT_REFS_DIR   = Path("").parent / "chip_refs"
SUPPORTED_EXTS     = {".png", ".jpg", ".jpeg", ".webp"}
CLIP_MODEL_ID      = "openai/clip-vit-base-patch32"   # ~600 MB, downloaded once
 
# Sliding-window crop settings
CROP_SIZES         = [200, 280, 360]   # square side lengths to try (px)
STRIDE_RATIO       = 0.4              # step = size * ratio  (0.4 = 60% overlap)
MATCH_THRESHOLD    = 0.80             # cosine similarity to count as a match
NMS_IOU_THRESHOLD  = 0.30             # suppress overlapping boxesa
```

```python
# ── Helpers ────────────────────────────────────────────────────────────────────
 
def pretty_name(filename: str) -> str:
    return Path(filename).stem.replace("-", " ").replace("_", " ").title()
 
 
def load_pil(path: Path) -> Image.Image:
    return Image.open(path).convert("RGB")
 
 
def cosine_similarity(a: torch.Tensor, b: torch.Tensor) -> torch.Tensor:
    a = a / a.norm(dim=-1, keepdim=True)
    b = b / b.norm(dim=-1, keepdim=True)
    return (a @ b.T)
 
 
def iou(box1, box2) -> float:
    ix1 = max(box1[0], box2[0]);  iy1 = max(box1[1], box2[1])
    ix2 = min(box1[2], box2[2]);  iy2 = min(box1[3], box2[3])
    inter = max(0, ix2 - ix1) * max(0, iy2 - iy1)
    area1 = (box1[2]-box1[0]) * (box1[3]-box1[1])
    area2 = (box2[2]-box2[0]) * (box2[3]-box2[1])
    union = area1 + area2 - inter
    return inter / union if union else 0.0
 
 
def nms(detections: list, iou_thresh: float) -> list:
    detections = sorted(detections, key=lambda d: d[0], reverse=True)
    kept = []
    for det in detections:
        if all(iou(det[2], k[2]) < iou_thresh for k in kept):
            kept.append(det)
    return kept
```

```python
# ── Main logic ─────────────────────────────────────────────────────────────────
 
def build_ref_embeddings(ref_paths, model, processor, device):
    print("📚  Encoding reference images ...")
    ref_labels, ref_embs = [], []
    for p in ref_paths:
        img = load_pil(p)
        inputs = processor(images=img, return_tensors="pt").to(device)
        with torch.no_grad():
            emb = model.get_image_features(**inputs)
        ref_labels.append(pretty_name(p.name))
        ref_embs.append(emb)
        print(f"   ✓ {pretty_name(p.name)}")
    ref_embs = torch.cat(ref_embs, dim=0)
    return ref_labels, ref_embs
 
 
def sliding_window_crops(image: np.ndarray, sizes: list, stride_ratio: float):
    h, w = image.shape[:2]
    for size in sizes:
        stride = max(1, int(size * stride_ratio))
        for y in range(0, h - size + 1, stride):
            for x in range(0, w - size + 1, stride):
                crop_np = image[y:y+size, x:x+size]
                crop_pil = Image.fromarray(cv2.cvtColor(crop_np, cv2.COLOR_BGR2RGB))
                yield (x, y, x+size, y+size, crop_pil)
 
 
def detect_chips(rack_image_path: Path, refs_dir: Path, threshold: float) -> None:
 
    # Load model
    print(f"\n🤖  Loading CLIP model ({CLIP_MODEL_ID}) ...")
    device = "cuda" if torch.cuda.is_available() else "cpu"
    print(f"    Device: {device}")
    model     = CLIPModel.from_pretrained(CLIP_MODEL_ID).to(device)
    processor = CLIPProcessor.from_pretrained(CLIP_MODEL_ID)
    model.eval()
 
    # Load references
    ref_paths = sorted([
        p for p in refs_dir.iterdir()
        if p.suffix.lower() in SUPPORTED_EXTS
    ])
    if not ref_paths:
        print(f"❌  No reference images found in: {refs_dir}")
        sys.exit(1)
 
    ref_labels, ref_embs = build_ref_embeddings(ref_paths, model, processor, device)
 
    # Load rack image
    rack_np = cv2.imread(str(rack_image_path))
    if rack_np is None:
        print(f"❌  Could not read rack image: {rack_image_path}")
        sys.exit(1)
    print(f"\n🔍  Scanning rack image: {rack_image_path.name}  "
          f"({rack_np.shape[1]}x{rack_np.shape[0]} px) ...")
 
    # Sliding-window matching
    detections = []
    crops_processed = 0
 
    for (x1, y1, x2, y2, crop_pil) in sliding_window_crops(rack_np, CROP_SIZES, STRIDE_RATIO):
        inputs = processor(images=crop_pil, return_tensors="pt").to(device)
        with torch.no_grad():
            crop_emb = model.get_image_features(**inputs)
 
        sims = cosine_similarity(crop_emb, ref_embs)[0]
        best_idx   = sims.argmax().item()
        best_score = sims[best_idx].item()
 
        if best_score >= threshold:
            detections.append((best_score, ref_labels[best_idx], (x1, y1, x2, y2)))
 
        crops_processed += 1
        if crops_processed % 200 == 0:
            print(f"   ... {crops_processed} crops checked")
 
    print(f"   Total crops checked: {crops_processed}")
 
    # Non-maximum suppression
    detections = nms(detections, NMS_IOU_THRESHOLD)
 
    # Print results
    print("\n" + "=" * 60)
    print("🛒  CHIPS DETECTED ON RACK")
    print("=" * 60)
 
    if not detections:
        print("  No chips matched above the similarity threshold.")
        print(f"  Try lowering --threshold (currently {threshold}).")
    else:
        counts: dict = {}
        for score, label, box in detections:
            counts.setdefault(label, []).append(score)
 
        for label, scores in sorted(counts.items()):
            avg = sum(scores) / len(scores)
            facings = len(scores)
            bar = "█" * min(facings, 10)
            print(f"  • {label:<30}  {facings:>2} facing(s)  "
                  f"[sim {avg:.2f}]  {bar}")
 
    print("=" * 60)
 
    # Save annotated image
    annotated = rack_np.copy()
    colors = [(0,200,0),(0,100,255),(255,100,0),(150,0,255),(0,200,200)]
    label_list = list({d[1] for d in detections})
    for score, label, (x1,y1,x2,y2) in detections:
        color = colors[label_list.index(label) % len(colors)]
        cv2.rectangle(annotated, (x1,y1), (x2,y2), color, 2)
        cv2.putText(annotated, label[:20], (x1, max(y1-6,10)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1, cv2.LINE_AA)
 
    out_path = rack_image_path.with_name(rack_image_path.stem + "_detected.jpg")
    cv2.imwrite(str(out_path), annotated)
    print(f"\n🖼   Annotated image saved -> {out_path}")
```

```python
# ── Entry point ────────────────────────────────────────────────────────────────
 
def main():
    parser = argparse.ArgumentParser(
        description="Detect chips on a rack — 100% local, no API key."
    )
    parser.add_argument("rack_image",
                        help="Path to the rack photo (jpg/png/webp)")
    parser.add_argument("--refs", default=str(DEFAULT_REFS_DIR),
                        help="Folder with reference images (default: chip_refs/)")
    parser.add_argument("--threshold", type=float, default=MATCH_THRESHOLD,
                        help=f"Cosine similarity cut-off (default: {MATCH_THRESHOLD})")
    args = parser.parse_args()
 
    rack_path = Path(args.rack_image)
    refs_path = Path(args.refs)
 
    if not rack_path.exists():
        print(f"❌  Rack image not found: {rack_path}")
        sys.exit(1)
    if not refs_path.exists():
        print(f"❌  Reference folder not found: {refs_path}")
        sys.exit(1)
 
    detect_chips(rack_path, refs_path, args.threshold)
```

```python
if __name__ == "__main__":
    main()
```
